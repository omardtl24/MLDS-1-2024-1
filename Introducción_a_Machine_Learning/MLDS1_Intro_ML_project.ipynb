```python
hd_data.select_dtypes(include=['int64', 'float64']).corr()
```

```python
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))


def cramers_v_matrix(df):
    columns = df.columns
    corr_matrix = pd.DataFrame(index=columns, columns=columns)
    for i in range(len(columns)):
        for j in range(len(columns)):
            if df[columns[i]].dtype == 'object' and df[columns[j]].dtype == 'object':
                corr_matrix.loc[columns[i], columns[j]] = cramers_v(df[columns[i]], df[columns[j]])
            else:
                corr_matrix.loc[columns[i], columns[j]] = np.nan
    return corr_matrix

cramers_v_correlation_matrix = cramers_v_matrix(hd_data.select_dtypes(include='object'))
cramers_v_correlation_matrix

```

### Modelo

Se realizará la implementación de algunos clasificadores usando las herramientas ofrecidas por las librerias Sklearn, Catboost y LightGBM. Se utilizarán puntualmente:

- *sklearn.ensemble.RandomForestClassifier:* Es un clasificador de aprendizaje automático basado en la técnica de bosques aleatorios, que combina múltiples árboles de decisión entrenados en diferentes muestras aleatorias de los datos de entrenamiento. Cada árbol emite una predicción y la clase más votada se selecciona como la predicción final del clasificador, lo que proporciona un modelo robusto y preciso al reducir el sobreajuste y mejorar la generalización gracias a la aleatoriedad introducida en la construcción de los árboles y la selección de características.

- *catboost.CatBoostClassifier:* Es un clasificador de aprendizaje automático desarrollado por Yandex, que se basa en el algoritmo de Gradient Boosting y está especialmente diseñado para manejar características categóricas en los datos de entrada. Utiliza un enfoque de boosting que entrena una serie de árboles de decisión secuenciales, donde cada árbol se enfoca en corregir los errores del anterior. Además, CatBoostClassifier implementa técnicas como el aprendizaje por refuerzo, la regularización y el manejo automático de características categóricas para mejorar la precisión del modelo y reducir el tiempo de entrenamiento.

-  *lightgbm.LGBMClassifier:* Es un clasificador de aprendizaje automático que se basa en el algoritmo de Gradient Boosting y está optimizado para velocidad y eficiencia en el manejo de grandes conjuntos de datos. Utiliza la técnica de boosting para construir un conjunto de árboles de decisión de manera secuencial, donde cada árbol se enfoca en corregir los errores del anterior. LGBMClassifier implementa una estrategia de división de hojas en lugar de división de nodos, lo que reduce significativamente el tiempo de entrenamiento al considerar solo las hojas que contienen datos relevantes. Esto lo hace especialmente efectivo en conjuntos de datos grandes y con muchas características.

Adicionalmente, para el preprocesamiento de los datos que serán ingresados al modelo, se utilizarán las siguientes herramientas:

- *StandardScaler():* El transformador StandardScaler es una técnica comúnmente utilizada para preprocesar características numéricas en un conjunto de datos. Su objetivo es transformar las características de tal manera que tengan una media igual a cero y una desviación estándar igual a uno. Esto significa que la distribución de cada característica estará centrada en cero y tendrá una dispersión comparable, lo que puede ser útil para algoritmos de aprendizaje automático que asumen que las características están en la misma escala.
- *OneHotEncoder():*  El transformador OneHotEncoder se utiliza para manejar características categóricas en un conjunto de datos. Convierte las características categóricas en un formato que los algoritmos de aprendizaje automático pueden entender más fácilmente, transformándolas en una representación binaria conocida como "codificación one-hot". Esto implica convertir cada categoría en una columna binaria separada, donde una columna se establece en 1 si la categoría está presente y 0 si no lo está. Esta codificación es crucial para evitar interpretaciones erróneas de las características categóricas como números ordinales.
- *ColumnTransformer:* La clase ColumnTransformer de sklearn es una herramienta poderosa para aplicar transformaciones específicas a columnas individuales o grupos de columnas en un conjunto de datos. Permite definir múltiples transformadores y aplicarlos selectivamente a diferentes subconjuntos de características en el conjunto de datos. Esto es útil cuando se trabaja con datos heterogéneos que requieren diferentes preprocesamientos según el tipo de características. Por ejemplo, se pueden aplicar transformaciones numéricas a características numéricas y transformaciones categóricas a características categóricas, todo dentro de una misma pipeline de procesamiento de datos. Esto facilita la construcción de flujos de trabajo de preprocesamiento de datos más complejos y flexibles en el aprendizaje automático.

```python
# Separando caracteristicas de variable objetivo
x = hd_data.drop('HeartDisease', axis=1)
y = hd_data['HeartDisease']
print(x.shape, y.shape)
```

Se utilizará una división estratificada para dividir el conjunto de datos en entrenamiento y prueba. Esta elección se justifica incluso cuando el desbalance en el conjunto de datos no es significativo, ya que asegura que tanto el conjunto de entrenamiento como el de prueba reflejen con precisión la distribución de clases del conjunto de datos original. Esto es crucial para una evaluación fiable del modelo en todas las clases y para prevenir el sobreajuste a la clase mayoritaria durante el entrenamiento. Además, promueve una generalización más efectiva del modelo a nuevos datos, mejorando su capacidad para enfrentar escenarios del mundo real.

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.3,
                                                    random_state=14,
                                                    stratify=y) # Se utiliza stratify debido a que es de interes mantener la proporción de datos para no caer en sesgo
```

```python
# Preprocesamiento con OneHotEncoder y  StandardScaler
categorical_features = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope', 'FastingBS']  # FastingBS se incluye debido a que realmente es categorica
numerical_features = ['Age', 'RestingBP', 'MaxHR', 'Oldpeak']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)
    ])
```

```python
# Pipeline y Selección de cada uno de los modelos
pipeline_sklearn = Pipeline([
                             ('preprocessor', preprocessor),
                             ('classifier', RandomForestClassifier(random_state=14))
                            ])
pipeline_catboost = Pipeline([
                             ('preprocessor', preprocessor),
                             ('classifier', CatBoostClassifier(random_seed=14))
                            ])
pipeline_lightgbm = Pipeline([
                             ('preprocessor', preprocessor),
                             ('classifier', LGBMClassifier(random_state=14))
                            ])

# Espacio de búsqueda de hiperparámetros
param_grid_sklearn = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [None, 10, 20, 30],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4]
}

param_grid_catboost = {
    'classifier__depth': [4, 6, 8],
    'classifier__learning_rate': [0.01, 0.05, 0.1],
    'classifier__iterations': [100, 200, 300]
}

param_grid_lgbm = {
    'classifier__num_leaves': [20, 30, 40],
    'classifier__max_depth': [10, 20, -1],
    'classifier__learning_rate': [0.01, 0.05, 0.1],
    'classifier__n_estimators': [100, 200, 300],
    'classifier__force_row_wise': [True, False],
    'classifier__force_col_wise': [True, False]
}

# Grilla para la búsqueda de hiperparámetros
grid_search_sklearn = GridSearchCV(pipeline_sklearn, param_grid_sklearn, cv=5, verbose=1)
grid_search_catboost = GridSearchCV(pipeline_catboost, param_grid_catboost, cv=5, verbose=1)
grid_search_lightgbm = GridSearchCV(pipeline_lightgbm, param_grid_lgbm, cv=5, verbose=1)
```

Relizamos el entrenamiento de los modelos
