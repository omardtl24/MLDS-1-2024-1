```python
primary_keys = {
    'teams':'TEAM_NAME',
    'stadiums': 'NAME',
    'players': 'PLAYER_ID',
    'matches': 'MATCH_ID',
    'goals' : 'GOAL_ID'

}

foreign_keys = {
    'teams': [{
        'key': 'HOME_STADIUM',
        'origin': 'stadiums(NAME)'
    }],
    'stadiums': [],
    'players': [{
        'key': 'TEAM',
        'origin': 'teams(TEAM_NAME)'
    }],
    'matches': [{
        'key': 'HOME_TEAM',
        'origin': 'teams(TEAM_NAME)'
    },{
        'key': 'AWAY_TEAM',
        'origin': 'teams(TEAM_NAME)'
    },{
        'key': 'STADIUM',
        'origin': 'stadiums(NAME)'
    }],
    'goals' : [{
        'key': 'MATCH_ID',
        'origin': 'matches(MATCH_ID)'
    },{
        'key': 'PID',
        'origin': 'players(PLAYER_ID)'
    },{
        'key': 'ASSIST',
        'origin': 'players(PLAYER_ID)'
    }]
}

data_types = {
    'object' : 'VARCHAR(50)',
    'int64' : 'INT',
    'datetime64[ns]': 'TIMESTAMP',
    'float64': 'FLOAT'
}

def convert_to_sql_timestamp(date_string):
  format_string = '%d-%b-%y %I.%M.%S.%f000000 %p'
  return datetime.datetime.strptime(date_string, format_string)

def foreign_constrains_query(name):
  table = name
  keys = foreign_keys[name]
  query =''
  for i in keys:
    subquery = f'''
          ALTER TABLE {name}
          ADD CONSTRAINT fk_{i['key'].lower()}
          FOREIGN KEY ({i['key']}) REFERENCES {i['origin']};

  '''
    query += subquery
  return query

def drop_table_query(name):

  return f"""
          DROP TABLE {name} CASCADE;
          """

def create_table_query(df,name):
  global primary_keys, data_types
  pk = primary_keys[name]
  columns = ', \n    '.join([str(i)+' '+data_types[str(df.dtypes[i])] for i in df.columns if i!=pk])
  query = f'''
  CREATE TABLE {name} (
    {pk +' '+data_types[str(df.dtypes[pk])]} PRIMARY KEY,
    {columns}
  );
  '''
  return query

def f_data(value):
  try:
    if np.isnan(value): return 'NULL'
    return str(value)
  except:
    if type(value)!=str:
      if value is np.nan: return 'NULL'
      if type(value)==pd._libs.tslibs.timestamps.Timestamp:
          return f" TIMESTAMP \'{value.strftime('%Y-%m-%d %H:%M:%S')}\'"
      if value is pd.NaT: return 'NULL'
      return str(value)
    else:
      value = value.replace("'", "''")
      return f"\'{value}\'"

def insert_query(df,name):
  columns = columns = ', '.join([str(i) for i in df.columns])
  query = f"""
  INSERT INTO {name} ({columns})
  VALUES
  {format_df_to_insert(df)}
  """
  return query

def format_df_to_insert(df):
  s = ''
  for _ , row in df.iterrows():
    r = ', '.join([f_data(i) for i in row])
    s += f"({r}),\n  "
  original_string = "Hello, World How are you?"
  pattern = r",\n  $"
  replacement = ";"
  s = re.sub(pattern, replacement, s)
  return s
```

```python
from scipy.stats import chi2_contingency

def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))


def cramers_v_matrix(df):
    columns = df.columns
    corr_matrix = pd.DataFrame(index=columns, columns=columns)
    for i in range(len(columns)):
        for j in range(len(columns)):
            if df[columns[i]].dtype == 'object' and df[columns[j]].dtype == 'object':
                corr_matrix.loc[columns[i], columns[j]] = cramers_v(df[columns[i]], df[columns[j]])
            else:
                corr_matrix.loc[columns[i], columns[j]] = np.nan
    return corr_matrix

df = df.select_dtypes(include='object')
df = df[['GOAL_DESC', 'SCORING_POSITION', 'ASSISTING_POSITION', 'SEASON', 'DATE_TIME']]
cramers_v_correlation_matrix = cramers_v_matrix(df)

cramers_v_correlation_matrix
```

```python
from pyspark.sql.functions import col, sum

# Function to check for null values in each column of a DataFrame
def check_missing_data(df, table_name):
    print(f"Missing data summary for table: {table_name}")
    missing_data = df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns])
    missing_data.show()

# Check missing data in each table
check_missing_data(df_teams, "teams")
check_missing_data(df_stadiums, "stadiums")
check_missing_data(df_players, "players")
check_missing_data(df_matches, "matches")
check_missing_data(df_goals, "goals")
```

```python
from pyspark.sql.functions import col, regexp_extract

# Function to check non-ASCII characters in each column of a DataFrame
def check_encoding_issues(df, table_name):
    print(f"Checking encoding issues for table: {table_name}")
    problematic_columns = []

    # Define a regular expression to detect non-ASCII characters
    non_ascii_regex = r'[^\x00-\x7F]'

    for column in df.columns:
        # Filter rows that contain non-ASCII characters
        problematic_rows = df.filter(regexp_extract(col(column), non_ascii_regex, 0) != "")
        non_ascii_count = problematic_rows.count()

        if non_ascii_count > 0:
            problematic_columns.append((column, non_ascii_count))
            print(f"Column: {column}, Non-ASCII Character Count: {non_ascii_count}")
            problematic_rows.show(truncate=False)

    if not problematic_columns:
        print(f"No encoding issues detected in table: {table_name}")

# Check encoding issues in each table
check_encoding_issues(df_teams, "teams")
print("-"*50)
check_encoding_issues(df_stadiums, "stadiums")
print("-"*50)
check_encoding_issues(df_players, "players")
print("-"*50)
check_encoding_issues(df_matches, "matches")
print("-"*50)
check_encoding_issues(df_goals, "goals")
```

```python
from pyspark.sql.functions import col

# Function to check duplicate elements in each column of a DataFrame
def check_duplicate_records(df, table_name):
    print(f"Duplicate records summary for table: {table_name}")
    total_records = df.count()
    unique_records = df.dropDuplicates().count()
    duplicate_records = total_records - unique_records
    print(f"Duplicate records: {duplicate_records}")


check_duplicate_records(df_teams, "teams")
check_duplicate_records(df_stadiums, "stadiums")
check_duplicate_records(df_players, "players")
check_duplicate_records(df_matches, "matches")
check_duplicate_records(df_goals, "goals")
```

# **Modelado**

Para realizar nuestra tarea de aprendizaje de máquina, tomaremos la información más importante de cada equipo y la organizaremos en un único dataframe que no dispone de valores nulos.

```python
matches_copy = df_matches.select("*")
players_copy = df_players.select("*")

#We get teams players average weight and height
players_copy = players_copy.select("team","height","weight")
players_avg = players_copy.groupby("team").agg(avg("height").alias("avg_height"),avg("weight").alias("avg_weight"))


#We get the teams match attendance proportion compared to stadium capacity average
matches_assist_prop  = matches_copy.join(df_stadiums, matches_copy.stadium == df_stadiums.name, "inner")
matches_assist_prop = matches_assist_prop.select("home_team", "attendance", "capacity")
matches_assist_prop = matches_assist_prop.withColumn("prop_atten", col("attendance") / col("capacity")).select("home_team", "prop_atten")
matches_assist_prop = matches_assist_prop.groupBy("home_team").agg(avg("prop_atten").alias("avg_prop_attendance"))

#We get the teams wins per match in home and away
matches_copy_h_a  = matches_copy.withColumn("home_win", col("home_team_score") > col("away_team_score")).select("home_team","away_team","home_team_score","away_team_score","home_win")
matches_copy_h_a  = matches_copy_h_a.withColumn("away_win", col("home_team_score") < col("away_team_score"))

matches_copy_h_a = matches_copy_h_a.withColumn("home_win_int", when(col("home_win") == True, 1).otherwise(0))
matches_copy_h_a = matches_copy_h_a.withColumn("away_win_int", when(col("away_win") == True, 1).otherwise(0))


matches_copy_h = matches_copy_h_a.select("home_team","home_team_score","away_team_score","home_win_int","away_win_int")
matches_copy_h = matches_copy_h_a.withColumnRenamed("away_win_int", "home_lost_int")
matches_copy_h = matches_copy_h.withColumn("DG_h", col("home_team_score") - col("away_team_score")).select("home_team", "DG_h","home_win_int","home_lost_int")
matches_copy_h = matches_copy_h.groupBy("home_team").agg(sum("DG_h").alias("DG_h"),sum("home_win_int").alias("win_h"),sum("home_lost_int").alias("lost_h"))



matches_copy_a = matches_copy_h_a.select("away_team","home_team_score","away_team_score","away_win_int","home_win_int")
matches_copy_a = matches_copy_h_a.withColumnRenamed("home_win_int", "away_lost_int")
matches_copy_a = matches_copy_a.withColumn("DG_a", col("away_team_score") - col("home_team_score")).select("away_team", "DG_a","away_win_int","away_lost_int")
matches_copy_a = matches_copy_a.groupBy("away_team").agg(sum("DG_a").alias("DG_a"),sum("away_win_int").alias("win_a"),sum("away_lost_int").alias("lost_a"))


matches_copy_h_a = matches_copy_h.join(matches_copy_a , matches_copy_h.home_team == matches_copy_a.away_team, "inner")
matches_copy_h_a = matches_copy_h_a.withColumnRenamed("home_team", "team")
matches_copy_h_a = matches_copy_h_a.withColumn("DG", col("DG_h") + col("DG_a")).select("team","DG","win_h","win_a","lost_h","lost_a")

#We join the three df into one single table by team
teams_info = matches_copy_h_a.join(matches_assist_prop , matches_copy_h_a.team == matches_assist_prop.home_team, "inner")
teams_info = teams_info.join(players_avg , players_avg.team == teams_info.home_team, "inner").select("home_team","DG","win_h","win_a","lost_h","lost_a","avg_prop_attendance","avg_height","avg_weight")
teams_info = teams_info.withColumnRenamed("home_team", "team")

#We show our df
teams_info.show()
```

```python
nulls_teams_info = teams_info.filter(
    col("team").isNull() |
    col("DG").isNull() |
    col("win_h").isNull() |
    col("win_a").isNull() |
    col("lost_h").isNull() |
    col("lost_a").isNull() |
    col("avg_prop_attendance").isNull() |
    col("avg_height").isNull() |
    col("avg_weight").isNull()
)

num_nulos = nulls_teams_info.count()
prop = 100*num_nulos/teams_info.count()
print(f'Info de equipos tiene un total de {num_nulos} filas con nulos, lo que representa un {round(prop,2)}% de la tabla')
```

Aplicamos un StandardScaler a los datos antes de ingresarlos al modelo de clustering

```python

numeric_columns = ['DG', 'win_h', 'win_a', 'lost_h', 'lost_a', 'avg_prop_attendance', 'avg_height', 'avg_weight']

assembler = VectorAssembler(inputCols=numeric_columns, outputCol="features")

df_assembled = assembler.transform(teams_info)

scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=False)

scaler_model = scaler.fit(df_assembled)
df_scaled = scaler_model.transform(df_assembled)

df_scaled = df_scaled.select("team", "scaled_features")
df_scaled.show(truncate=False)
```

```python
k_values = range(2, 11)

inertia_values = []

for k in k_values:
    kmeans = KMeans(featuresCol="scaled_features", k=k)
    model = kmeans.fit(df_scaled)
    summary = model.summary
    inertia = summary.trainingCost
    inertia_values.append(inertia)

plt.plot(k_values, inertia_values, marker='o')
plt.xlabel('Número de Clústeres (K)')
plt.ylabel('Inercia')
plt.title('Método del Codo para encontrar el valor óptimo de K')
plt.grid(True)
plt.show()
```

Dado el resultado, podemos determinar que $k=5$ representa el codo de la gráfica y es por consiguiente el valor adecuado para el clustering. Procedemos a entrenar el modelo con este hiperparametro.

```python
kmeans = KMeans(featuresCol="scaled_features", k=5)
model = kmeans.fit(df_scaled)
clustered = model.transform(df_scaled).select("*")
clustered.show()
```

```python
clusters = clustered.groupBy("prediction").agg(collect_list("team").alias("teams_in_cluster")).orderBy("prediction")
clusters.show()
```

# **Análisis de resultados**

```python

pca = PCA(k=3, inputCol="scaled_features", outputCol="pca_features")
pca_model = pca.fit(clustered)
df_pca = pca_model.transform(clustered)
df_pca = df_pca.withColumnRenamed('prediction','Cluster')

pandas_df = df_pca.select("Cluster", "pca_features").toPandas()


pandas_df[['pca_1', 'pca_2', 'pca_3']] = pandas_df['pca_features'].apply(lambda x: pd.Series(x))

fig = go.Figure()

for cluster in clusters:
    cluster_data = pandas_df[pandas_df["Cluster"] == cluster]
    fig.add_trace(go.Scatter3d(
        x=cluster_data['pca_1'],
        y=cluster_data['pca_2'],
        z=cluster_data['pca_3'],
        mode='markers',
        name=f'Cluster {cluster}',
        marker=dict(
            size=5,
            opacity=0.8
        )
    ))


fig.update_layout(scene=dict(
                    xaxis_title='Componente principal 1',
                    yaxis_title='Componente principal 2',
                    zaxis_title='Componente principal 3'),
                  title='PCA con 3 componentes')

fig.show()
```
